Seaborn

```python
tips=pd.read_csv("/content/tips.csv")
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.stripplot(y="tip",data=tips,jitter=True)
plt.ylabel("tips")
plt.show()
```

```python
sns.stripplot(x="day",y="tip",data=tips)
plt.ylabel("tips")
plt.show()
```

```python
sns.stripplot(x="day",y="tip",data=tips,jitter=False)
plt.show()
```

```python
sns.swarmplot(x="day",y="tip",data=tips)
plt.show()
```

```python
sns.swarmplot(x="day",y="tip",hue="sex",data=tips)
plt.show()
```

```python
sns.boxplot(x="day",y="tip",data=tips)
plt.show()
```

```python
sns.violinplot(x="day",y="tip",data=tips)
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="day",y="tip",data=tips,color="red")
plt.subplot(1,2,2)
sns.violinplot(x="day",y="tip",data=tips)
plt.show()
```

```python
sns.violinplot(x="day",y="tip",data=tips,inner=None,color="lightgray")
sns.stripplot(x="day",y="tip",data=tips,size=4,jitter=True)
plt.show()
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,color="green")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,color="green",kind="kde")
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(hue="sex",data=tips)
```
